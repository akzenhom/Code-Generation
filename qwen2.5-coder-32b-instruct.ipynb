````python
def decisionEngine(income, category):
    minIncome = 2500
    minIncomeMoh = 3500
    minIncomeStudent = 4500
    maxIncome = 6500
    maxIncomeStudent = 5500
    minIncomeMilitaryStudent = 3520
    minIncomeMilitaryOfficer = 8520
    minIncomeMilitaryNonOfficer = 8562

    if category == "MOH":
        if income >= minIncomeMoh and income <= maxIncome:
            return "Eligible"
        else:
            return "Not Eligible"
    elif category == "Student":
        if income >= minIncomeStudent and income <= maxIncomeStudent:
            return "Eligible"
        else:
            return "Not Eligible"
    elif category == "MilitaryStudent":
        if income >= minIncomeMilitaryStudent and income <= maxIncome:
            return "Eligible"
        else:
            return "Not Eligible"
    elif category == "MilitaryOfficer":
        if income >= minIncomeMilitaryOfficer:
            return "Eligible"
        else:
            return "Not Eligible"
    elif category == "MilitaryNonOfficer":
        if income >= minIncomeMilitaryNonOfficer:
            return "Eligible"
        else:
            return "Not Eligible"
    else:
        if income >= minIncome and income <= maxIncome:
            return "Eligible"
        else:
            return "Not Eligible"
````
